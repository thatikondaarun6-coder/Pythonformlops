```python
# f"string {expression}" syntax of fsting
# MLOPS EXAMPLES FOR F STRING FORMAT
#Example 1: Logging ML Metrics
#
model_name = "RandomForest"
accuracy = 0.92345
loss = 0.123456
print(f"Model: {model_name:<15} | Accuracy: {accuracy:.2%} | Loss: {loss:.4f}")
models = ["RandomForest", "XGBoost", "SVM"]
for m in models:
    print(f"Model: {m:<15} | Status: Ready")

```

```python
#{task:^20} → Formatting instruction:means center-align the text means reserve 20 character spaces: equal spaces on both sides.<20 → left-align

#:>20 → right-align

#:^20 → center-align

task = "Data Preprocessing"
status = "SUCCESS"
duration = 123.456

print(f"Task: {task:^20} | Status: {status:<10} | Duration: {duration:.2f}s")
```

```python
# Numeric formatting which deals numbers
# syntax f"{value:[alignment][width][.precision][type]}"
pi = 3.1415926535
print(f"{pi:.3f}")   # 2 decimal places

```

```python
accuracy = 0.8765
print(f"{accuracy:.2%}")   # 2 decimal places as %
```
