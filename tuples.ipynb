```python
# Accessing Tuple Elements and slicing

fruits = ("apple", "banana", "cherry")
print(fruits[0])   # Output: apple
print(fruits[1])   # Output: banana

# Negative indexing
print(fruits[-1])  # Output: cherry (last element)

numbers = (0, 1, 2, 3, 4, 5)
print(numbers[1:3])
print(numbers[2:4])
print(numbers[:4])
print(numbers[2:])

```
