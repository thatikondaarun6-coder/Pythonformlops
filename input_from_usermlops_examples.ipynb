```python
#Model Deployment Configuration with User Confirmation
# deploy_model.py
import json

def load_and_confirm_config(config_path):
    with open(config_path) as f:
        cfg = json.load(f)

    print("Configuration loaded:")
    print(json.dumps(cfg, indent=2))
    confirm = input("\nProceed with deployment? (y/n): ")

    if confirm.lower() == "y":
        print("Deploying model...")
        return True
    else:
        print("Deployment cancelled.")
        return False

if __name__ == "__main__":
    config_file = input("Enter config file path: ")
    if load_and_confirm_config(config_file):
        # Deploy logic here
        print("Deployment successful!")
```

```python
#Data Validation Script with User Input

# validate_data.py
import os

def validate_dataset():
    data_path = input("Enter dataset path: ")

    if not os.path.exists(data_path):
        print("Error: File not found!")
        return False

    test_size = float(input("Enter test split ratio (0.0-1.0): "))
    if test_size < 0 or test_size > 1:
        print("Error: Invalid test size!")
        return False

    print(f"✓ Dataset validation passed")
    print(f"  Path: {data_path}")
    print(f"  Test split: {test_size * 100}%")
    return True

if __name__ == "__main__":
    validate_dataset()
```
