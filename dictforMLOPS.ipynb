```python
dicv = student = [{
    "name": "John",
    "age": 20,
    "grade": "A",
    "subject": "Python",
    "score": 95
},
                  {
    "name": "Arun",
    "age": 22,
    "grade": "A+",
    "subject": "Python",
    "score": 95
}]

print("All values from all dictionaries in the list:")
for record_dict in dicv: # Iterate through each dictionary in the list
  for key, value in record_dict.items(): # Get both key and value from the current dictionary
    print(f"{key}: {value}")
```
